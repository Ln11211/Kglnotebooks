```python
x=data_train.drop('churn',axis=1)
y=data_train['churn']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from category_encoders.target_encoder import TargetEncoder
from xgboost import XGBClassifier

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15)

steps=[('enc',TargetEncoder()),('clf',XGBClassifier())]

pipe=Pipeline(steps)
```

```python
pipe
```

```python
from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer

search_space = {
    'clf__max_depth': Integer(1,10),
    'clf__max_leaves': Integer(1,10),
    'clf__n_estimators': Integer(250,1000),
    'clf__learning_rate': Real(0.001, 1, prior='log-uniform'),
    'clf__subsample': Real(0.5, 1.0),
    'clf__colsample_bytree': Real(0.5, 1.0),
    'clf__colsample_bylevel': Real(0.5, 1.0),
    'clf__colsample_bynode' : Real(0.5, 1.0),
    'clf__reg_alpha': Real(0.2, 10.0),
    'clf__reg_lambda': Real(0.2, 10.0),
    'clf__gamma': Real(0.2, 10.0)
}

opt = BayesSearchCV(pipe, search_space, cv=5, n_iter=5, scoring='roc_auc') 
# in reality, you may consider setting cv and n_iter to higher values
```

```python
opt.fit(x_train,y_train)
```

```python
opt.best_estimator_
```

```python
opt.best_score_
```

```python
opt.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error

pre=opt.predict(x_test)
mae=mean_absolute_error(pre,y_test)
mae
```

```python
from xgboost import plot_importance

xgboost_step = opt.best_estimator_.steps[1]
xgboost_model = xgboost_step[1]
plot_importance(xgboost_model)
```

```python
from sklearn.ensemble import RandomForestClassifier

steps=[('enc',TargetEncoder()),('tr',RandomForestClassifier())]

pipe_tree=Pipeline(steps)
pipe_tree
```

```python
pipe_tree[1].get_params(deep=True)
```

```python
search_space = {
    'tr__max_depth': Integer(1,10),
    'tr__max_leaf_nodes': Integer(2,10),
    'tr__ccp_alpha': Real(0.001, 1),
    'tr__max_depth': Integer(1,10),
    'tr__max_leaf_nodes': Integer(1,20),
    'tr__max_samples': Integer(1,20),
    'tr__min_samples_leaf': Integer(1,20),
    'tr__min_samples_split': Integer(1,20),
    'tr__n_estimators': Integer(500,1500),}

tree = BayesSearchCV(pipe_tree, search_space, cv=5, n_iter=7, scoring='roc_auc')
```

```python
tree.fit(x_train,y_train)
```

```python
tree.best_estimator_
```

```python
tree.best_score_
```

```python
tree.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error

pre=tree.predict(x_test)
mae=mean_absolute_error(pre,y_test)
mae
```
